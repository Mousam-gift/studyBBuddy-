```python
from google import genai
from google.genai import types

client = genai.Client()

SYSTEM_PROMPT = """
You are a helpful assistant that can help in three areas:
1) Education: explain concepts, make study plans, create quizzes.
2) Health information: ONLY general health info and lifestyle tips, no diagnosis or prescriptions. Always say you are not a doctor.
3) Sustainability: give practical eco-friendly habits and low-carbon tips.

Always:
- Mention which area(s) you are using (Education / Health info / Sustainability).
- Keep answers concise and clear.
"""

def smart_agent(message: str) -> str:
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[
            types.Content(role="user", parts=[types.Part(text=SYSTEM_PROMPT)]),
            types.Content(role="user", parts=[types.Part(text=message)]),
        ],
    )
    return response.text
```

```python
def classify_domain(message: str) -> str:
    m = message.lower()
    if any(w in m for w in ["study", "exam", "homework", "learn", "subject"]):
        return "education"
    if any(w in m for w in ["health", "diet", "exercise", "sleep", "doctor", "symptom"]):
        return "health"
    if any(w in m for w in ["climate", "carbon", "sustainable", "eco", "recycle", "footprint"]):
        return "sustainability"
    return "mixed"

def smart_routed_agent(message: str) -> str:
    domain = classify_domain(message)
    prefixed = f"[DOMAIN={domain}] {message}"
    return smart_agent(prefixed)
```

```python
print(smart_agent("Help me make a weekly study plan that is healthy and sustainable."))
```

```python
print(smart_routed_agent("Explain recursion and give me 3 quiz questions."))
print(smart_routed_agent("What habits improve heart health?"))
print(smart_routed_agent("How to reduce my carbon footprint as a student?"))
```
